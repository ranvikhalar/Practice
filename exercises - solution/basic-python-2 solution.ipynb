```python
import numpy as np
import pandas as pd

#Ques 1.  Imports the data into a data structure of your choice
#Ques 2.  Loops over the rows in the dataset
#Ques 3.  For each row in the dataset checks to see if the ear length is large (>10 cm) or small (<=10 cm) and determines the GC-content of the DNA sequence (i.e., the percentage of bases that are either G or C)
#Ques 4.  Stores this information in a table where the first column has the ID for the individual, the second column contains the string 'large' or the string 'small' depending on the size of the individuals ears, and the third column contains the GC content of the DNA sequence.
#Ques 5.  Prints the average GC-content for both large-eared elves and small-eared elves to the screen.
#Ques 6.  Exports the table of individual level GC values to a CSV (comma delimited text) file titled `grangers_analysis.csv`.

def get_data_from_web(url):
    data = pd.read_csv(url)

    table = []
    large_gc = []
    small_gc = []
    for i in range(len(data)):
        id = data.iloc[i]['id']
        ear_length = data.iloc[i]['earlength']
        DNA_sequence = data.iloc[i]['dnaseq']
        gc_count = DNA_sequence.count("G") + DNA_sequence.count("C")
        gc_content = (gc_count / len(DNA_sequence)) * 100
        if ear_length > 10:
            length = 'large'
            large_gc.append(gc_content)
        else:
            length = 'small'
            small_gc.append(gc_content)        
        
        table.append([id, length, gc_content])

    new_data = pd.DataFrame(table, columns = ['ID', 'length', 'gc_content'])

    new_data.to_csv('grangers_analysis.csv', index=False)
    print(f"Average GC-content for large-eared elves = {np.mean(large_gc)}")
    print(f"Average GC-content for small-eared elves = {np.mean(small_gc)}")
        
    return new_data


get_data_from_web('C:/Users/15199/courses/data/houseelf_earlength_dna_data.csv')

```
